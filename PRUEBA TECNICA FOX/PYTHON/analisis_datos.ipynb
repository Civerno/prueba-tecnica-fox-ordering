```python
import pandas as pd
import numpy as np
import openpyxl as px
import pyarrow as pq
import json as js

MaestroClientes = pd.read_parquet("PD_Maestro_Clientes.parquet")
VentasMes = pd.read_excel('PD_Ventas_Mes.xlsx')
MaestroProductos = pd.read_json('PD_Maestro_productos.json')

```

```python
print("=== MastroClientes ===")
print(MaestroClientes.info())
```

```python
print("=== VentasMes ===")
print(VentasMes.head(10))
print(VentasMes.info())
```

```python
print("=== MaestroProductos ===")
print(type(MaestroProductos))
print(MaestroProductos.head(10))
print(MaestroProductos.info())
```

```python
#VISTA MINABLE

VistaMinable = VentasMes.merge(MaestroProductos, on='ItemId', how='left')
VistaMinable = VistaMinable.merge(MaestroClientes, on='id_cliente', how='left')

print("=== VistaMinable ===")
print(VistaMinable.head(10))    
print(VistaMinable.info())
```

```python
#MISING VALUES

print(VistaMinable.isnull().sum())

# % MISIN VALUES

print((VistaMinable.isnull().sum()/len(VistaMinable))*100)

```

```python
#ESTADISTICAS DESCRIPTIVAS

print(VistaMinable.describe())

#CORRELACIONES

correlaciones = VistaMinable.select_dtypes(include=['int64', 'float64']).corr()
print("=== Correlaciones ===")
print(correlaciones)
```
